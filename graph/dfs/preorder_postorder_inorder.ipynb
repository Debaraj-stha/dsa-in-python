```python
class Node:
    def __init__(self, value=0):
        self.value = value
        self.left = None
        self.right = None

def inorder(node):
    if node:
        inorder(node.left)
        print(node.value, end=' ')
        inorder(node.right)
def inorderByDFS(node):
    stack=[]
    current=node
    while current or stack:
        while current:
            stack.append(current)
            current=current.left
        current=stack.pop()
        print(current.value,end=' ')
        current=current.right

def preorder(node):
    if node:
        print(node.value, end=' ')
        preorder(node.left)
        preorder(node.right)


def preorderDFS(node):
    stack=[]
    current=node
    while current or stack:
        while current:
            print(current.value, end=' ')
            stack.append(current)
            current=current.left
        current=stack.pop()
        current=current.right

def postorder(node):
    if node:
        postorder(node.left)
        postorder(node.right)
        print(node.value, end=' ')


def postorderDFS(node):
    stack = []
    visited = set() 
    
    current = node
    while current or stack:
        while current:
            stack.append(current)
            current = current.left  
            
        current = stack[-1]
        if not current.right or current.right in visited:  
            print(current.value, end=' ') 
            visited.add(current)  
            stack.pop()  
            current = None 
        else:
            current = current.right 


            
        
       
        


root=Node(4)
root.left=Node(2)
root.right=Node(5)
root.left.left=Node(1)
root.left.right=Node(3)
inorder(root)
print()
inorderByDFS(root)
print()
preorder(root)
print()
preorderDFS(root)
print()
postorder(root)
print()
postorderDFS(root)
```
