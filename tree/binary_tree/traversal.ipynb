```python
from collections import  defaultdict,deque
class Node:
    def __init__(self,value):
        self.value = value
        self.left = None
        self.right = None

def preorder(root):
    if root:
        print(root.value, end=' ')
        preorder(root.left)
        preorder(root.right)

def preorderiterative(root):
    if not root:
        return
    stack = deque()
    stack.append(root)
    while stack:
        node = stack.pop()
        print(node.value, end=' ')
        if node.right:
            stack.append(node.right)
        if node.left:
            stack.append(node.left)

def inorder(root):
    if root:
        inorder(root.left)
        print(root.value, end=' ')
        inorder(root.right)

def inorderiterative(root):
    ans = []
    stack = []
    curr = root

    while curr is not None or len(stack) > 0:
        while curr is not None:
            stack.append(curr)
            curr = curr.left
        curr = stack.pop()
        ans.append(curr.value)
        curr = curr.right

    return ans


def postOrder(root):
    if root:
        postOrder(root.left)
        postOrder(root.right)
        print(root.value, end=' ')

def postOrderiterative(root):
    if not root:
        return
   
    stack1 = deque()
    stack2 = deque()
    stack1.append(root)
    while stack1:
        node = stack1.pop()
        stack2.append(node.value)
        if node.left:
            stack1.append(node.left)
        if node.right:
            stack1.append(node.right)
    while stack2:
        print(stack2.pop(), end=' ')


def levelOrderTraversal(root):
    if not root:
        return
    queue = deque()
    queue.append(root)
    while queue:
        node=queue.popleft()
        print(node.value, end=' ')
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)

def reverseLevelOrderTraversal(root):
    if not root:
        return
    queue=deque([root])
    stack2=[]
    while queue:
        node=queue.popleft()
        stack2.append(node)
        if node.right:
            queue.append(node.right)
        if node.left:
            queue.append(node.left)
    while stack2:
        print(stack2.pop().value, end=' ')
    


def zigzagTraversal(node):
    queue=deque([node])
    level=0
    res=[]
    while queue:
        size=len(queue)
        levelNodes=[]
        for _ in range(size):
            temp=queue.popleft()
            levelNodes.append(temp.value)
            if temp.left:
                queue.append(temp.left)
            if temp.right:
                queue.append(temp.right)
        if level%2==1:
            levelNodes.reverse()
        res.extend(levelNodes)
        level+=1
    return res

def morrisTraversal(root):
    res=[]
    curr=root
    while curr is not None:
        if curr.left is None:

            # If no left child, visit this node 
            # and go right
            res.append(curr.value)
            curr = curr.right
        else:

            # Find the inorder predecessor of curr
            prev = curr.left
            while prev.right is not None   and prev.right != curr:
                prev = prev.right

            # Make curr the right child of its 
            # inorder predecessor
            if prev.right is None:
                prev.right = curr
                curr = curr.left
            else:

                # Revert the changes made in the 
                # tree structure
                prev.right = None
                res.append(curr.value)
                curr = curr.right

    return res

        


node=Node(44)
node.left=Node(22)
node.right=Node(66)
node.left.left=Node(11)
node.left.right=Node(33)
node.right.left=Node(55)
node.right.right=Node(77)
print("preorder traversal")
preorder(node)
print("")
preorderiterative(node)
print("\ninorder")
inorder(node)
print("")
print(preorderiterative(node))
print("\npostorder")
postOrder(node)
print()
postOrderiterative(node)
print("\nlevel order traversal")
print(levelOrderTraversal(node))
print("\nzigzag traversal")
print(zigzagTraversal(node))
print("reverse level order traversal=")
reverseLevelOrderTraversal(node)
```
