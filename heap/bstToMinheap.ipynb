```python
import heapq
from collections import deque

class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None

def inorder(node):
    if not node:
        return []
    nodes = []
    stack = []
    current = node
    while stack or current:
        while current:
            stack.append(current)
            current = current.left
        current = stack.pop()
        nodes.append(current.key)
        current = current.right
    return nodes

def levelOrder(node):
    if not node:
        return []
    queue = deque([node])
    while queue:
        current = queue.popleft()
        print(current.key, end=' ')
        if current.left:
            queue.append(current.left)
        if current.right:
            queue.append(current.right)

def bstToMinheap(root):
    if not root:
        return None
    inorder_nodes = inorder(root)
    min_heap=[]
    for node_key in inorder_nodes:
        heapq.heappush(min_heap, node_key)
    def buildHeap(node):
        if not node:
            return
        node.key=heapq.heappop(min_heap)
        buildHeap(node.left)
        buildHeap(node.right)

    buildHeap(root)
    return root
def preorder(root):
    if not root:
        return
    print(root.key, end=' ')
    preorder(root.left)
    preorder(root.right)

def inorder_morris(root):
    result = []
    current = root

    while current:
        if current.left is None:
            result.append(current.key)
            current = current.right
        else:
            # Find the inorder predecessor of current
            pre = current.left
            while pre.right and pre.right is not current:
                pre = pre.right

            if pre.right is None:
                pre.right = current  # Make a temporary link
                current = current.left
            else:
                pre.right = None  # Restore tree structure
                result.append(current.key)
                current = current.right

    return result


root = Node(4)
root.left = Node(2)
root.right = Node(6)
root.left.left = Node(1)
root.left.right = Node(3)
root.right.left = Node(5)
root.right.right = Node(7)

print("BST Inorder Traversal:")
print(inorder(root))
print("bst level order Traversal:")
levelOrder(root)
print("\npreorder bst")
preorder(root)

nr = bstToMinheap(root)

print("\nMin-Heap Inorder Traversal:")
print(inorder(nr))

print("Min-Heap Level Order Traversal:")
levelOrder(nr)


print("\npreorder minheap")
preorder(nr)
```
