```python
class Node:
    def __init__(self, value=0):
        self.value = value
        self.next = None

def display(node):
    current = node
    visited = set()
    
    while current:
        if current in visited:
            print(f'({current.value})', end=' -> LOOP DETECTED')
            return
        print(current.value, end=' -> ')
        
        visited.add(current)  # Add before moving to next node
        current = current.next  # Move to the next node
    
    print('(head)')  # End of a normal linked list


def swap_pairs(node):
    if not node:
        return None
    
    dummy = Node(0)
    dummy.next = node
    prev = dummy
    
    while prev.next and prev.next.next:
        first = prev.next
        second = first.next

        # Swap nodes
        prev.next = second
        first.next = second.next
        second.next = first

        # Move prev forward
        prev = first

    return dummy.next  # Return the new head

def reverse(node):
    if not node:
        return None  

    prev = None
    current = node
    
    while current:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
    
    return prev  # New head of reversed list

def reverse_until(node, k):
    if not node or k <= 0:
        return None, node

    prev = None
    count = 0
    current = node

    while current and count < k:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
        count += 1
    
    return prev, current  # Return new head and remaining list


def cycle(head,index):
    loopNode = None
    pos = 0
    current = head
    lastNode = None  # To store the last node

    while current:
        if pos == index:
            loopNode = current 
        lastNode = current  
        current = current.next
        pos += 1

    if lastNode and loopNode:
        print("Loop created at node:", loopNode.value)
        lastNode.next = loopNode 



def has_cycle(head):
    slow = fast = head
    while fast and fast.next:
       
        slow = slow.next
        fast = fast.next.next
       
        if slow == fast:
            
            startNode = head
            index = 0
            while startNode != slow:
                startNode = startNode.next
                slow = slow.next
                index += 1
            return True, index
    return False, -1

def remove_cycle(node):
    if not node:
        return None
    has_loop, index=has_cycle(node)
    if not has_loop:
        print("No loop found")
        return node
    if index==-1:
        print("Invalid loop index")
        return node
    print(has_loop,index)
    
    if has_loop:
        current=node
        for _ in range(index):
           current=current.next
        loop_start=current
        while current.next!=loop_start:
           current=current.next
        current.next = None
       
    


# Create the linked list: 3 -> 5 -> 8 -> 9 -> 10 -> 11
node = Node(3)
node.next = Node(5)
node.next.next = Node(8)
node.next.next.next = Node(9)
node.next.next.next.next = Node(10)
node.next.next.next.next.next = Node(11)

print("Original List:")
display(node)

cycle(node,2)
display(node)

# Step 4: move the cycle
rem = remove_cycle(node)
print("After Removing Cycle:")
display(node)
```
