```python
class Node:
    def __init__(self, value=0):
        self.value = value
        self.next = None
        self.prev = None  

def display(head):
    """ Display a circular singly linked list. """
    if not head:
        print("Empty list")
        return

    current = head
    while True:
        print(current.value, end=" -> ")
        current = current.next
        if current == head:
            break
    print("(head)")

def toDoublyCircularLinkedList(head):
    """ Convert singly circular linked list to doubly circular linked list. """
    if not head:
        return None

    current = head
    prev = None
    first = head 

    while True:
        current.prev = prev  
        prev = current 
        current = current.next  

        if current == first:  
            break

    first.prev = prev  
    prev.next = first  

    return head

def displayDoublyCircularLinkedList(head):
    """ Display a doubly circular linked list in both directions. """
    if not head:
        print("Empty list")
        return

    print("Forward: ", end="")
    current = head
    while True:
        print(current.value, end=" <-> ")
        current = current.next
        if current == head:
            break
    print("(head)")

    print("Reverse: ", end="")
    tail = head.prev  
    current = tail
    while True:
        print(current.value, end=" <-> ")
        current = current.prev
        if current == tail:
            break
    print("(tail)")


node = Node(4)
node.next = Node(2)
node.next.next = Node(1)
node.next.next.next = Node(33)
node.next.next.next.next = Node(55)
node.next.next.next.next.next = node  

print("Original Circular Singly Linked List:")
display(node)


dcll = toDoublyCircularLinkedList(node)

print("\nConverted Doubly Circular Linked List:")
displayDoublyCircularLinkedList(dcll)
```
