```python
class Node:
    def __init__(self,value=0):
        self.value = value
        self.next = None
def mergeSort(node):
    if not node or not node.next:
        return node
    slow=node
    fast=node.next.next
    while fast and fast.next:
        slow=slow.next
        fast=fast.next.next
    mid=slow.next
    slow.next=None
    left=mergeSort(node)
    right=mergeSort(mid)
    return merge(left,right)
def merge(left,right):
    if not left:
        return right
    if not right:
        return left
    if left.value<=right.value:
        result=left
        result.next=merge(left.next,right)
    if right.value<=left.value:
        result=right
        result.next=merge(left,right.next)
    return result


def insertionSort(node):
    def sortedInsert(sortedHead,newNode):
        if not sortedHead:
            sortedHead=newNode
            return sortedHead
        if newNode.value<=sortedHead.value:
            newNode.next=sortedHead
            return newNode
        current=sortedHead
        while current.next and current.next.value < newNode.value:
            current=current.next
        newNode.next=current.next
        current.next=newNode
        return sortedHead
    if not node:
        return node
    current=node
    sortedHead=None
    while current:
            next_node=current.next
            current.next=None
            sortedHead=sortedInsert(sortedHead, current)
            current=next_node
    return sortedHead


def selectionSort(node):
    if not node:
        return
    current = node
    while current:
        minNode = current
        search=current.next
        while search:
            if search.value<minNode.value:
                minNode=search
            search=search.next
        current.value,minNode.value=minNode.value,current.value
        current=current.next
    return node


def quicksort(node):
    if not node or not node.next:
        return node
    tail=node #find tail node
    while tail.next:
        tail=tail.next
    node=quicksortHelper(node,tail) #reasign updated node
    return node
def partition(start,end):
    if not start or not end:
        return start
    pivot=end
    i=start
    j=start
    while j!=pivot:
        if j.value<=pivot.value:#swap if smaller than pivot
            i.value, j.value=j.value, i.value
            i=i.next
        j=j.next
    i.value, pivot.value=pivot.value, i.value
    return i

def quicksortHelper(start, end):
    if not start or start == end:
        return start

    # Partition returns the pivot node
    pivot = partition(start, end)

    # Find node before pivot
    prev = None
    curr = start
    while curr != pivot:
        prev = curr
        curr = curr.next

    # Recurse on left part (before pivot)
    if prev:
        prev.next = None  # Temporarily break list
        sorted_left = quicksortHelper(start, prev)

        # Find tail of sorted_left to connect with pivot
        tail = sorted_left
        while tail.next:
            tail = tail.next
        tail.next = pivot
    else:
        sorted_left = pivot  # Pivot is the first element

    # Recurse on right part (after pivot)
    pivot.next = quicksortHelper(pivot.next, end)

    return sorted_left


# Testing the function
head = Node(5)
head.next = Node(2)
head.next.next = Node(8)
head.next.next.next = Node(1)
head.next.next.next.next = Node(4)
head.next.next.next.next.next=Node(3)

sort = quicksort(head)

# Print the sorted linked list
current = sort
while current:
    print(current.value, end=' ')
    current = current.next
print()
# Output: 1 2 4 5 8
```
