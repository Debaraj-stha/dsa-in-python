```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def insert(self, value):
        new_node = Node(value)
        if not self.head:
            self.head = new_node
            self.head.next = self.head
            return
        last = self.head
        while last.next != self.head:
            last = last.next
        last.next = new_node
        new_node.next = self.head

    def partition(self, start, end):
        """Partition function for QuickSort, rearranging nodes properly."""
        if start == end or start is None or end is None:
            return start

        pivot_value = end.value
        i = start
        j = start

        while j != end:
            if j.value < pivot_value: 
                i.value, j.value = j.value, i.value
                i = i.next
            j = j.next

        i.value, end.value = end.value, i.value
        return i  

    def quicksort_recursive(self, start, end):
        if start is None or start == end or end is None:
            return

        pivot = self.partition(start, end)

        prev = start
        while prev.next != pivot:
            prev = prev.next

        # Recursively sort left and right parts
        self.quicksort_recursive(start, prev)
        self.quicksort_recursive(pivot.next, end)

    def quicksort(self):
        """Wrapper function to find last node and call recursive QuickSort."""
        if self.head is None:
            return

        last = self.head
        while last.next != self.head:
            last = last.next

        self.quicksort_recursive(self.head, last)

    def display(self):
        if not self.head:
            print("List is empty")
            return
        current = self.head
        visited = set()
        while current not in visited:
            print(current.value, end=" -> ")
            visited.add(current)
            current = current.next
        print("(head)")

    def bubblesort(self):
        isSorted=False
        while not isSorted:
            isSorted=True
            current=self.head
            while current.next!=self.head:
                next_node=current.next
                if current.value>next_node.value:
                    current.value,next_node.value=next_node.value,current.value
                    isSorted=False
                current=current.next


   


# **Test Case**

cll=CircularLinkedList()
cll.insert(3)
cll.insert(1)
cll.insert(5)
cll.insert(2)
cll.insert(4)
# cll.quicksort()
cll.insertionsort()
cll.display()
```
