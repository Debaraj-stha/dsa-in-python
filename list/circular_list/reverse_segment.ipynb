```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyCircularLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            new_node.next = new_node
        else:
            current = self.head
            while current.next != self.head:
                current = current.next
            current.next = new_node
            new_node.next = self.head

    def display(self):
        if not self.head:
            print("List is empty")
            return
        current = self.head
        while True:
            print(current.data, end=" -> ")
            current = current.next
            if current == self.head:
                break
        print("...")

    def reverse_segment(self, start_index, end_index):
        if not self.head:
            return

        if start_index >= end_index:
            return
        prev_start = None
        current = self.head
        index = 0
        while index <= start_index - 1:
            prev_start = current
            current = current.next
            index += 1

        if start_index > 0:
            start = prev_start.next
        else:
            start = self.head

        end = start
        index = start_index
        while index < end_index:
            end = end.next
            index += 1
        after_end=end.next
        current=start
        prev=after_end
        while current!=after_end:
            next_node=current.next
            current.next=prev
            prev=current
            current=next_node
        if start_index>0:
            prev_start.next=end
        else:
            self.head=end
        start.next=after_end


# Example usage:
scll = SinglyCircularLinkedList()
scll.append(1)
scll.append(2)
scll.append(3)
scll.append(4)
scll.append(5)
scll.append(6)
scll.append(7)

print("Original list:")
scll.display()

scll.reverse_segment(2, 5)

print("List after reversing segment (indices 2 to 5):")
scll.display()
```
