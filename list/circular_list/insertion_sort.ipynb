```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def insert(self, value):
        """Inserts a node at the end of the circular linked list."""
        new_node = Node(value)
        if not self.head:
            self.head = new_node
            self.head.next = self.head
            return
        last = self.head
        while last.next != self.head:
            last = last.next
        last.next = new_node
        new_node.next = self.head

    def display(self):
        """Displays the circular linked list."""
        if not self.head:
            print("List is empty")
            return
        current = self.head
        first = True
        while first or current != self.head:
            first = False
            print(current.value, end=" -> ")
            current = current.next
        print("(head)")

    def sorted_insert(self, sorted_head, new_node):
        """Inserts an existing node into the correct sorted position."""
        if not sorted_head: 
            new_node.next = new_node
            return new_node

        if new_node.value < sorted_head.value:
            last = sorted_head
            while last.next != sorted_head: 
                last = last.next
            last.next = new_node
            new_node.next = sorted_head
            return new_node 
        current = sorted_head
        while current.next != sorted_head and current.next.value < new_node.value:
            current = current.next

        new_node.next = current.next
        current.next = new_node
        return sorted_head

    def insertionsort(self):
        """Sorts the circular linked list using insertion sort."""
        if not self.head or self.head.next == self.head:
            return  

        sorted_head = None  
        current = self.head
        while True:
            next_node = current.next  
            current.next = None  
            sorted_head = self.sorted_insert(sorted_head, current) 
            current = next_node 

            if current == self.head:  
                break

        self.head = sorted_head 

    def selectionSort(self):
        if not self.head or self.head.next == self.head:
            return
        
        current = self.head
        while True:
            minNode = current
            search=current.next
            while search!=self.head:
                if search.value < minNode.value:
                    minNode = search
                search = search.next
            current.value,minNode.value=minNode.value,current.value
            current = current.next
            if current==self.head:
                break

    

# Test case
cll = CircularLinkedList()
cll.insert(3)
cll.insert(1)
cll.insert(5)
cll.insert(2)
cll.insert(4)

print("Original List:")
cll.display()

cll.isort()
# cll.selectionSort()
print("Sorted List:")
cll.display()
```
