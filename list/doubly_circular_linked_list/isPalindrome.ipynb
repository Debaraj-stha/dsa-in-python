```python
class Node(object):
    def __init__(self,value=0):
        self.value = value
        self.prev=None
        self.next=None
     


def displayFromHead(head):
    current = head
    while True:
        print(current.value, end='<->')
        current = current.next
        if current==head:
            break
    print("tail")

def displayFromTail(tail):
    current = tail
    while True:
        print(current.value, end='<->')
        current = current.prev
        if current==tail:
            break
    print("head")




def isPalindrome(head,tail):
    if head is None or head==tail:
        return True

    origiona_head=head
    origional_tail=tail
    while head!=tail:
        if head.value!=tail.value:
            return False,origiona_head,origional_tail
        head=head.next
        tail=tail.prev
    head=origiona_head   
    tail=origional_tail 
    return True,head,tail

node=Node(1)
node1=Node(9)
node2=Node(8)
node3=Node(91)
node4=Node(8)
node5=Node(91)
node6=Node(8)
node7=Node(97)
node8=Node(1)



head=node
tail=node8

node.next=node1
node1.next=node2
node2.next=node3
node3.next=node4
node4.next=node5
node5.next=node6
node6.next=node7
node7.next=node8
node8.next=node


node1.prev=node
node2.prev=node1
node3.prev=node2
node4.prev=node3
node5.prev=node4
node6.prev=node5
node7.prev=node6
node8.prev=node7
node.prev=node8





print("from head")
displayFromHead(head)
print("from tail")
displayFromTail(tail)
ispalind,nhead,ntail=isPalindrome(head,tail)
print("is palindrome",ispalind)
displayFromHead(head)
displayFromTail(tail)
```
