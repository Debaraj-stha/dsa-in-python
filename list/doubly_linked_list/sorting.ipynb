```python
from node import Node
from display import display, displayFromTail

def get_middle(head):
    """Finds the middle node of the doubly linked list."""
    if not head or not head.next:
        return head

    slow, fast = head, head
    while fast.next and fast.next.next:
        slow = slow.next
        fast = fast.next.next

    return slow

def merge(left, right):
    """Merges two sorted halves of a doubly linked list."""
    if not left:
        return right
    if not right:
        return left

    dummy = Node(0)
    current = dummy
    while left and right:
        if left.value <= right.value:
            current.next = left
            left.prev = current  # Maintain prev links
            left = left.next
        else:
            current.next = right
            right.prev = current  # Maintain prev links
            right = right.next
        current = current.next

    # Attach the remaining part
    if left:
        current.next = left
        left.prev = current
    if right:
        current.next = right
        right.prev = current

    new_head = dummy.next
    new_head.prev = None  # Ensure head's prev is None
    return new_head


def merge_sort(head):
    """Performs an iterative bottom-up merge sort on a doubly linked list."""
    if not head or not head.next:
        return head

    # Split list into two halves
    middle = get_middle(head)
    right_half = middle.next
    middle.next = None
    if right_half:
        right_half.prev = None

    # Recursively sort each half
    left_sorted = merge_sort(head)
    right_sorted = merge_sort(right_half)

    # Merge sorted halves
    return merge(left_sorted, right_sorted)


def bubblesort(head):
    isSorted=False
    while not isSorted:
        isSorted=True
        current = head
        while current.next:
            next_node = current.next
            if current.value > next_node.value:
                current.value, next_node.value = next_node.value, current.value
                isSorted = False
            current = current.next
def partition(head,tail):
    if not head or not tail or head == tail:
        return head

    pivot = tail
    i = head
    j = head

    while j != tail:
        if j.value <= pivot.value:
            i.value, j.value = j.value, i.value
            i = i.next
        j = j.next

    i.value, pivot.value = pivot.value, i.value
    return i
def quickSort(head, tail):
    if not head or not tail or head == tail or head == tail.next:
        return

    pivot = partition(head, tail)

    quickSort(head, pivot.prev)  # Sort left partition
    quickSort(pivot.next, tail)  # Sort right partition



node = Node(44)
node1 = Node(4)
node2 = Node(1)
node3 = Node(2)
node4 = Node(3)
node5 = Node(5)
node6 = Node(6)
node7 = Node(7)

node.next = node1
node1.prev = node
node1.next = node2
node2.prev = node1
node2.next = node3
node3.prev = node2
node3.next = node4
node4.prev = node3
node4.next = node5
node5.prev = node4
node5.next = node6
node6.prev = node5
node6.next = node7
node7.prev = node6

head = node
tail=node7

print("Original List (from head):")
display(head)
print("\nOriginal List (from tail):")
displayFromTail(node7)
print()
quickSort(head,tail)
display(head)
# bubblesort(head)
# print("bubblesort")
# display(head)
# Sort the list
# head = merge_sort(head)

# Find new tail after sorting
# tail = head
# while tail.next:
#     tail = tail.next

# print("After Merge Sort (from head):")
# display(head)
# print("\nAfter Merge Sort (from tail):")
# displayFromTail(tail)
```
