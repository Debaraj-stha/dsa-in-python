```python
class Node(object):
    def __init__(self,value=0):
        self.value = value
        self.prev=None
        self.next=None
        self.child=None


def displayFromHead(head):
    current = head
    while current:
        print(current.value, end='<->')
        current = current.next
    print("None")

def displayFromTail(tail):
    current = tail
    while current:
        print(current.value, end='<->')
        current = current.prev
    print("None")


def displayChild(head):
    current = head
    while current:
        print(current.value, end='<->')
        child=current.child
        while child:
            print(child.value, end='<->')
            child=child.child
        current=current.next
    print("None")

def flatten(head):
    if head is None:
        return 
    dummy=Node()
    ref=dummy
    current=head
    while current:
        ref.child=Node(current.value)
        ref=ref.child
        child=current.child
        while child:
            ref.child=Node(child.value)
            ref=ref.child
            child=child.child
        current=current.next
    return dummy.child

node=Node(1)
node1=Node(9)
node2=Node(8)
node3=Node(91)
node4=Node(22)
node5=Node(33)
node6=Node(11)
node7=Node(21)

#children

child1=Node(88)
child2=Node(99)
child3=Node(6)
child4=Node(22)
child5=Node(33)

head=node
tail=node7

node.next=node1
node1.next=node2
node2.next=node3
node3.next=node4
node4.next=node5
node5.next=node6
node6.next=node7


node1.prev=node
node2.prev=node1
node3.prev=node2
node4.prev=node3
node5.prev=node4
node6.prev=node5
node7.prev=node6


node1.child=child1
child1.child=child2
child2.child=child3


node2.child=child4
child4.child=child5


print("from head")
displayFromHead(head)
print("from tail")
displayFromTail(tail)
print("with child")
displayChild(head)
flatten=flatten(head)
print("flattened list")
displayChild(flatten)
# while flatten:
#     print(flatten.value, end='<->')
#     flatten=flatten.child
# print("None")
```
