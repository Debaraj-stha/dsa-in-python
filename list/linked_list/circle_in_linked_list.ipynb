```python
class Node:
    def __init__(self, value=0):
        self.value = value
        self.next = None

def display(head):
    visited = set()  # To prevent infinite loops
    while head:
        if head in visited:
            print(f"({head.value})", end=" -> LOOP DETECTED")
            return
        visited.add(head)
        print(head.value, end=" -> ")
        head = head.next
    print("None")

def crate_loop(index, head):
    loopNode = None
    pos = 0
    current = head
    lastNode = None  # To store the last node

    while current:
        if pos == index:
            loopNode = current 
        lastNode = current  
        current = current.next
        pos += 1

    if lastNode and loopNode:
        print("Loop created at node:", loopNode.value)
        lastNode.next = loopNode 

def has_cycle(head):
    slow = fast = head
    while fast and fast.next:
        print("slow:", slow.value,"fast ",fast.value)
        slow = slow.next
        fast = fast.next.next
       
        if slow == fast:
            print(slow.value,"fast ",fast.value)
            startNode = head
            index = 0
            while startNode != slow:
                print("slow:", slow.value, "start:", startNode.value)
                startNode = startNode.next
                slow = slow.next
                index += 1
            return True, index
    return False, -1, None

def deleteLoop(head):
    has_loop, index = has_cycle(head)
    if has_loop:
        loopStart = head
        for _ in range(index):  # Move loopStart to loop entrance
            loopStart = loopStart.next

        current = loopStart
        while current.next != loopStart:
            current = current.next
        current.next = None  # Break the loop
        return "Loop deleted successfully"
    else:
        return "No loop found"

head = Node("A")
head.next = Node("B")
head.next.next = Node("C")
head.next.next.next = Node("D")
head.next.next.next.next = Node("E")
head.next.next.next.next.next = Node("F")
head.next.next.next.next.next.next = Node("G")
print("Before creating loop:")
display(head)
crate_loop(2, head)
print("After creating loop:")
display(head)
print(deleteLoop(head))

print("After removing loop:")
display(head)
```
