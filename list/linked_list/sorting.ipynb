```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def sorting(head,reverse=False):
    isSorted=False
    while not isSorted:
        isSorted=True
        current=head
        while current.next:
            next_node=current.next
            if (reverse and current.data < next_node.data) or (not reverse and current.data > next_node.data):
                current.data, next_node.data = next_node.data, current.data
                isSorted=False
            current=current.next
    return head
def selectionSort(head):
    if not head:
        return None
    current=head
    while current.next:
        minNode=current
        searchNode=current.next
        while searchNode:
            if minNode.data > searchNode.data:
                minNode=searchNode
            searchNode=searchNode.next
        current.data, minNode.data=minNode.data, current.data
        current=current.next
    return head

def partition(start,end):
    if not start:
        return end
    if not end:
        return start
    pivot=end
    i=start
    j=start
    while j!=pivot:
        if j.data < pivot.data:
            i.data, j.data=j.data, i.data
            i=i.next
        j=j.next
    i.data, pivot.data=pivot.data, i.data
    return i

def quickSort(start, end):
    if not start or not end or start==end:
        return start
    pivot=partition(start, end)
    prev=start
    while prev.next!=pivot:
        prev=prev.next
    quickSort(start, prev)
    quickSort(pivot.next, end)
    return start
def sorted_insert(sorted_head,newNode):
    if not sorted_head:
        sorted_head=newNode
        return sorted_head
    if newNode.data <= sorted_head.data:
        newNode.next=sorted_head
        return newNode
    current=sorted_head
    while current.next and current.next.data < newNode.data:
        current=current.next
    newNode.next=current.next
    current.next=newNode
    return sorted_head
def insertionSort(head):
    if not head:
        return head
    sorted_head=None
    current=head.next
    while current:
        nextNode=current.next
        current.next=None
        sorted_head = sorted_insert(sorted_head, current)
        current=nextNode
    return sorted_head

def split(head):
    if not head or not head.next:
        return head, None
    slow=head
    fast=head.next
    while fast and fast.next:
        slow=slow.next
        fast=fast.next.next
    mid=slow.next
    slow.next=None
    return head, mid
def merge(left,right):
    if not left:
        return right
    if not right:
        return left
    dummy=Node(0)
    current=dummy
    while left and right:
        if left.data < right.data:
            current.next=left
            left=left.next
        else:
            current.next=right
            right=right.next
        current=current.next
    current.next=left if left else right
    return dummy.next
def mergesort(start,end):
    if not start or not end or start==end:
        return start
    first,second=split(start)
    last=first
    while last.next:
        last=last.next
    left=mergesort(start,last)
    right=mergesort(second, end)
    return merge(left,right)



# Test the function
node1=Node(3)
node2=Node(5)
node3=Node(8)
node4=Node(1)
node5=Node(2)
node6=Node(9)

node1.next=node2
node2.next=node3
node3.next=node4
node4.next=node5
node5.next=node6

head=node1

print("Original Linked List:")
current = head
while current:
    print(current.data, end=" ")
    current = current.next

# new_head = sorting(head)
# new_head=selectionSort(head)

# print("\nSorted Linked List:")
# current = new_head
# while current:
#     print(current.data, end=" ")
#     current = current.next

# new_head = quickSort(head, node6)
new_head=mergesort(head,node6)

print("\nQuick Sorted Linked List:")
current = new_head
while current:
    print(current.data, end=" ")
    current = current.next
```
