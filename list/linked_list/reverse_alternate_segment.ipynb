```python
from node import  Node
from display import  display,displayFromTail
def reverseAlternateSegment(head,start,end):
    dummy=Node(0)
    dummy.next=head
    prev=dummy
    current=head
    for _ in range(start):
        prev=current
        current=current.next
    segment_start=current
    segment_prev=prev
    for _ in range(end-start+1):
        next_node=current.next
        current.next=prev
        prev=current
        current=next_node
    segment_prev.next=prev
    segment_start.next=current
    head=dummy.next
    return head


def length(head):
    count=0
    current=head
    while current:
        count+=1
        current=current.next
    return count

def reverseAlternate(head, k):
    l = length(head)
    if l < 2 or k <= 0:
        return head

    segmentStart = 0
    current_head = head
    flip = True 

    while segmentStart < l:
        if flip:
            segmentEnd = min(segmentStart + k - 1, l - 1)
            current_head = reverseAlternateSegment(current_head, segmentStart, segmentEnd)
        segmentStart += k
        flip = not flip 
    return current_head


node = Node(44)
node1 = Node(4)
node2 = Node(1)
node3 = Node(2)
node4 = Node(3)
node5 = Node(5)
node6 = Node(6)
node7 = Node(7)
node.next = node1
node1.next = node2
node2.next = node3
node3.next = node4
node4.next=node5
node5.next=node6
node6.next=node7
node7.next=Node(8)
node7.next.next=Node(9)
node7.next.next.next=Node(10)
display(node)
print()
nhead=reverseAlternate(node,3)
display(nhead)
```
