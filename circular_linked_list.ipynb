```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
        
```

```python
class CircularLinkedList:
    def __init__(self,value):
        new_node = Node(value)
        self.head = new_node
        new_node.next = new_node 
    def display(self):
        if not self.head:
            print("List is empty.")
            return
        
        temp = self.head
        while True:
            print(temp.value, end=" -> ")
            temp = temp.next
            if temp == self.head:
                break
        print("(back to head)")

    def insertAtFirst(self, value):
        new_node = Node(value)
        current = self.head
        
        while current.next != self.head:
            current = current.next
        
        new_node.next = self.head
        current.next = new_node
        self.head = new_node 

    def insertAtLast(self, value):
        new_node=Node(value)
        current=self.head
        while current.next!=self.head:
            current=current.next
        current.next=new_node
        new_node.next=self.head


    def length(self):
        count=0
        current=self.head
        while current.next!=self.head:
            count+=1
            current=current.next
        return count+1
    
    def insertAt(self,index,value):
        if index<0 or index>self.length():
            raise IndexError("Index out of bounds")
        new_node=Node(value)
        current=self.head
        prev=None
        for _ in range(index):
            prev=current
            current=current.next
        new_node.next=current
        prev.next=new_node

    def deleteFirst(self):
        current=self.head
        if current==self.head.next:
            self.head=None
        while current.next!=self.head:
            current=current.next
        current.next=self.head.next
        self.head=self.head.next

    def deleteLast(self):
        current=self.head
        while current.next.next!=self.head:
            current=current.next
        current.next=self.head.next
        self.head=self.head.next

    def deleteAt(self,index):
        if index<0 or index>=self.length():
            raise IndexError("Index out of bounds")
        if index==0:
            self.deleteFirst()
            return
        elif index==self.length()-1:
            self.deleteLast()
            return
        prev=None
        current=self.head
        for _ in range(index):
            prev=current
            current=current.next
        prev.next=current.next
    def inert(self,*args):
        for value in args:
            self.insertAtLast(value)

    def reverse(self):
        current=self.head
        prev=None
        while True:
            next_node=current.next
            current.next=prev
            prev=current
            current=next_node
            if current==self.head:
                break
        self.head.next=prev
        self.head=prev
        
    


    



      
```

```python
cll=CircularLinkedList(4)

cll.insertAtFirst(3)
cll.insertAtFirst(2)
cll.display()
cll.insertAtLast(43)
cll.insertAtLast(42)
cll.display()
print("Length of the Circular Linked List is:", cll.length())
cll.insertAt(3, 35)
cll.display()
cll.deleteLast()
cll.display()
cll.deleteFirst()
cll.display()
cll.deleteAt(1)
cll.display()
cll.inert(1,2,3,5,6,7,8,9)
cll.display()
cll.reverse()
cll.display()
```
