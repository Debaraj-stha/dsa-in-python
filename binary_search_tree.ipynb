```python
class Node:
    def __init__(self,value=None):
        self.value = value
        self.left=None
        self.right=None
    
```

```python
class BinarySearchTree:
    def __init__(self,node=None):
        self.root=node
    def insert(self,value,node):
        if value<node.value:
            if node.left is None:
                node.left=Node(value)
            else:
                self.insert(value,node.left)
        else:
            if node.right is None:
                node.right=Node(value)
            else:
                self.insert(value,node.right)

    def delete(self,value,node):
        if value<node.value:
            node.left=self.delete(value,node.left)
        elif value>node.value:
            node.right=self.delete(value,node.right)
        else:
            if node.left is None:
                temp=node.right
                return temp
            elif node.right is None:
                temp=node.left
                return temp
            temp=self._minvalue_node(node.right)
            node.value=temp.value
            node.right=self.delete(temp.value,node.right)
        return node

      
    
      
    def _minvalue_node(self,node):
        current=node
        while current.left:
            current=current.left
        return current



    def inorder(self,node):
        if node:
            self.inorder(node.left)
            print(node.value, end=" ")
            self.inorder(node.right)

    def preorder(self,node):
        if node:
            print(node.value, end=" ")
            self.preorder(node.left)
            self.preorder(node.right)

    def postorder(self,node):
        if node:
            self.postorder(node.left)
            self.postorder(node.right)
            print(node.value,end=" ")

    def height(self,node):
        if not node:
            return 0
        left=self.height(node.left)
        right=self.height(node.right)
        return 1+max(left,right)
    def find_min(self,node):
        if not node:
            return float('-inf')
        return self._minvalue_node(node).value
    def find_max(self,node):
        if not node:
            return float('inf')
        current=node
        while current.right:
            current=current.right
        return current.value
    

    def search(self, value, node):
            if node is None or node.value==value:
                return node
            if value<node.value:
                return self.search(value,node.left)
            return self.search(value,node.right)
    

    def isBalanced(self, node):
        if  node is None:
            return True
        left_height=self.height(node.left)
        right_height=self.height(node.right)
        if abs(left_height-right_height)<=1 and self.isBalanced(node.left) and self.isBalanced(node.right):
            return True
        return False
    
    def predecessor(self,node,key):
        predecessor=None
        while node:
            if key>node.value:
                predecessor=node #update predecessor when moving right
                node=node.right
            elif key<node.value:
                node=node.left
            else:
                if node.left:
                    temp=node.left
                    while temp.right:
                        temp=temp.right
                    return temp
                break
        return predecessor
    
    def successor(self, node, key):
        successor=None
        while node:
            if key>node.value:
                node=node.right
            elif key<node.value:
                successor=node #update successor when moving left
                node=node.left

            else:
                if node.right:
                    temp=node.right
                    while temp.left:
                        temp=temp.left
                    return temp
                break
            
        return successor
    def parent(self, node, key):
        if node is None or (node.left is None and node.right is None):
            return None  

        if (node.left and node.left.value == key) or (node.right and node.right.value == key):
            return node  

        if key < node.value:
            return self.parent(node.left, key)  
        else:
            return self.parent(node.right, key)  
    def left_children(self,node):
        children=[]
        temp=self.search(node.value,node)
        if temp and temp.left:
            self.collect_children(temp.left,children)
        return children
    def collect_children(self,node,children):
        if node:
            children.append(node.value)
            self.collect_children(node.left,children)
            self.collect_children(node.right,children)
    def right_children(self,node):
        temp=self.search(node.value,node)
        children=[]
        if temp and temp.right:
            self.collect_children(temp.right,children)
        return children


    
    def kthSmallest(self, node, k):
            if node is None:
                return None
            count=self.countNode(node.left)
            if count==k-1:
                return node.value
            if count<k-1:
                return self.kthSmallest(node.right,k-count-1)
            return self.kthSmallest(node.left,k)
    def countNode(self,node):
        if node is None:
            return 0
        return 1+self.countNode(node.left)+self.countNode(node.right)
    
    def kthLargest(self,node,k):
        if node is None:
            return None
        count=self.countNode(node.right)
        if count==k-1:
            return node.value
        if count<k-1:
            return self.kthLargest(node.left,k-count-1)
        return self.kthLargest(node.right,k)
    def isValidBST(self, node, min_val=float('-inf'), max_val=float('inf')):
        if node is None:
            return True
        if node.value <= min_val or node.value >= max_val:
            return False
        return (self.isValidBST(node.left, min_val, node.value) and
                self.isValidBST(node.right, node.value, max_val))
    def lowestCommonAncestor(self, node, p, q):
        print("called")
        if node is None:
            return None

        # If both p and q are smaller, LCA is in the left subtree
        if p < node.value and q < node.value:
            return self.lowestCommonAncestor(node.left, p, q)

        # If both p and q are greater, LCA is in the right subtree
        elif p > node.value and q > node.value:
            return self.lowestCommonAncestor(node.right, p, q)

        # If split occurs, node is the LCA
        return node

    def distanceBetween(self,node,p,q):
        lca=self.lowestCommonAncestor(node,p,q)
        return self.depth(lca,p,0)+self.depth(lca,q,0)
    def depth(self,node,target,level):
        if node is None:
            return -1
        if node.value==target:
            return level
        left=self.depth(node.left,target,level+1)
        return left if left!=-1 else self.depth(node.right,target,level+1)
    

    def serialize(self, node):
        def preorder(node):
            return [str(node.value)] + preorder(node.left) + preorder(node.right) if node else ["#"]
        return " ".join(preorder(node))
    def deserialize(self, data):
        def build():
            val = next(values)
            if val == "#":
                return None
            node = Node(int(val))
            node.left = build()
            node.right = build()
            return node

        values = iter(data.split())
        return build()
```
